A permutation perm of n + 1 integers of all the integers in the range [0, n] can be represented as a string s of length n where:

- s[i] == 'I' if perm[i] < perm[i + 1], and
- s[i] == 'D' if perm[i] > perm[i + 1].

Given a string s, reconstruct the permutation perm and return it. If there are multiple valid permutations perm, return **any of them**.

**Example 1:**

**Input:** s = "IDID"

**Output:**

[0,4,1,3,2]

</aside>

```python
def reconstruct_permutations(s):
  perm=[]
  n=len(s)
  low=0
  high=n
  for character in s:
    if character == 'I':
      perm.append(low)
      low=low+1
    elif character =='D':
      perm.append(high)
      high=high-1

  perm.append(low)
  return perm

```

```python
s = "IDID"
result=reconstruct_permutations(s)
print(result)
```

You are given an m x n integer matrix matrix with the following two properties:

- Each row is sorted in non-decreasing order.
- The first integer of each row is greater than the last integer of the previous row.

Given an integer target, return true *if* target *is in* matrix *or* false *otherwise*.

You must write a solution in O(log(m * n)) time complexity.

**Example 1:**

</aside>**Input:** matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], target = 3

**Output:** true
image.png

```python
def find_target_in_sorted_matrix(matrix, target):
  """
  Finds the target in the sorted matrix.

  Args:
    matrix: The sorted matrix.
    target: The target value to search for.

  Returns:
    True if the target is found, False otherwise.
  """

  low = 0
  high = len(matrix) * len(matrix[0]) - 1

  while low <= high:
    mid = (low + high) // 2

    if matrix[mid // len(matrix)][mid % len(matrix[0])] == target:
      return True
    elif matrix[mid // len(matrix)][mid % len(matrix[0])] < target:
      low = mid + 1
    else:
      high = mid - 1

  return False


if __name__ == "__main__":
  matrix = [[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]]
  target = 3

  if find_target_in_sorted_matrix(matrix, target):
    print("Target found")
  else:
    print("Target not found")
```

Given a binary array nums, return *the maximum length of a contiguous subarray with an equal number of* 0 *and* 1.

**Example 1:**

**Input:** nums = [0,1]

**Output:** 2

**Explanation:**

[0, 1] is the longest contiguous subarray with an equal number of 0 and 1.

</aside>

```python
def findMaxLength(nums):
    count = 0
    max_length = 0
    count_dict = {0: -1}  # Initialize the dictionary with a count of 0 at index -1

    for i in range(len(nums)):
        # Increment count by 1 for 1 and decrement by 1 for 0
        count += 1 if nums[i] == 1 else -1
        
        # If the current count is already in the dictionary, update the max length
        if count in count_dict:
            max_length = max(max_length, i - count_dict[count])
        else:
            count_dict[count] = i

    return max_length

```

```python
 nums = [0,1]
 result=findMaxLength(nums)
 print(result)
```

The **product sum** of two equal-length arrays a and b is equal to the sum of a[i] * b[i] for all 0 <= i < a.length (**0-indexed**).

- For example, if a = [1,2,3,4] and b = [5,2,3,1], the **product sum** would be 1*5 + 2*2 + 3*3 + 4*1 = 22.

Given two arrays nums1 and nums2 of length n, return *the **minimum product sum** if you are allowed to **rearrange** the **order** of the elements in* nums1.

**Example 1:**

**Input:** nums1 = [5,3,4,2], nums2 = [4,2,2,5]

**Output:** 40

**Explanation:**

We can rearrange nums1 to become [3,5,4,2]. The product sum of [3,5,4,2] and [4,2,2,5] is 3*4 + 5*2 + 4*2 + 2*5 = 40.

</aside>

```python
a = [1,2,3,4]
b = [5,2,3,1]
result=min_product_sum(a,b)
print(result)
```

```python
def min_product_sum(a,b):
  a.sort()
  b.sort(reverse=True)#Sort it descending order
  n=len(a)
  minimum_product_sum=0
  for i in range(n):
    minimum_product_sum=a[i]*b[i]

  return minimum_product_sum

```

An integer array original is transformed into a **doubled** array changed by appending **twice the value** of every element in original, and then randomly **shuffling** the resulting array.

Given an array changed, return original *if* changed *is a **doubled** array. If* changed *is not a **doubled** array, return an empty array. The elements in* original *may be returned in **any** order*.

**Example 1:**

**Input:** changed = [1,3,4,2,6,8]

**Output:** [1,3,4]

**Explanation:** One possible original array could be [1,3,4]:

- Twice the value of 1 is 1 * 2 = 2.
- Twice the value of 3 is 3 * 2 = 6.
- Twice the value of 4 is 4 * 2 = 8.

Other original arrays could be [4,3,1] or [3,1,4].

</aside>

```python
def original_array(changed):
  if not all(ele %2==0 for ele in changed ):
    return [] 
  # Create a map to store the elements of changed and their counts.
  counts={}
  for ele in changed:
    counts=counts.get(ele,0)+1
  #create a list to store original array
  original=[]
  #Iterate over the elements of changed and add them to the original array if their count is 1.
  for ele in changed:
    if counts[ele]==1:
      original.append(ele)

  return original
```

Given a positive integer n, generate an n x n matrix filled with elements from 1 to n2 in spiral order.

**Example 1:**

**Input:** n = 3

**Output:** [[1,2,3],[8,9,4],[7,6,5]]

</aside>

```python
def generateMatrix(n):
    matrix = [[0] * n for _ in range(n)]  # Initialize the matrix with zeros
    num = 1  # Start with the number 1
    row_start, row_end = 0, n - 1
    col_start, col_end = 0, n - 1

    while row_start <= row_end and col_start <= col_end:
        # Traverse right
        for j in range(col_start, col_end + 1):
            matrix[row_start][j] = num
            num += 1
        row_start += 1

        # Traverse down
        for i in range(row_start, row_end + 1):
            matrix[i][col_end] = num
            num += 1
        col_end -= 1

        # Traverse left
        if row_start <= row_end:
            for j in range(col_end, col_start - 1, -1):
                matrix[row_end][j] = num
                num += 1
            row_end -= 1

        # Traverse up
        if col_start <= col_end:
            for i in range(row_end, row_start - 1, -1):
                matrix[i][col_start] = num
                num += 1
            col_start += 1

    return matrix
```

</aside>

```python
mat1 = [[1,0,0],[-1,0,3]]
mat2 = [[7,0,0],[0,0,0],[0,0,1]]
# Multiply the matrices.
result = sparse_matrix_multiplication(mat1, mat2)

# Print the result matrix.
print(result)
```

```python
def sparse_matrix_multiplication(mat1, mat2):
    m, k, n = len(mat1), len(mat1[0]), len(mat2[0])
    result = [[0] * n for _ in range(m)]
    mat2_sparse = [[(j, mat2[j][col]) for col in range(n) if mat2[j][col] != 0] for j in range(k)]

    for i in range(m):
        for j in range(n):
            for col, value in mat2_sparse[j]:
                result[i][j] += mat1[i][col] * value

    return result
```
